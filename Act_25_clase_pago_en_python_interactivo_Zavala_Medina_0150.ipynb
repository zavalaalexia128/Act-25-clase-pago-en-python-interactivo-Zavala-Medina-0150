### Crear la clase Pago, con 3 atributos (monto, fecha, método)
### 2 funciones leerdatos(), mostrardatos() y 2 objetos (pago1, pago2).
### que sea interctivo --> input
### no usar while true, if, for, lista y diccionarios

## El pago
![](https://raw.githubusercontent.com/zavalaalexia128/Act-25-clase-pago-en-python-interactivo-Zavala-Medina-0150/refs/heads/main/infopagos.png)


```python
# Clase Pago, atributos: monto, fecha, metodo
# funciones leerdatos(), mostrardatos()
# crear 2 objetos pago1, pago2

class Pago:
    def __init__(self, monto=None, fecha=None, metodo=None):
        self.monto = monto
        self.fecha = fecha
        self.metodo = metodo

    def leerdatos(self):
        self.monto = input("Ingrese el monto del pago: ")
        self.fecha = input("Ingrese la fecha (dd/mm/aaaa): ")
        self.metodo = input("Ingrese el método de pago (Efectivo/Tarjeta): ")
        return self.monto, self.fecha, self.metodo

    def mostrardatos(self):
        print(f"Monto: ${self.monto}")
        print(f"Fecha: {self.fecha}")
        print(f"Método de pago: {self.metodo}")
        return self.monto, self.fecha, self.metodo

print("--- Registro de Pago Numero 1 ---")
pago1 = Pago()
pago1.leerdatos()
pago1.mostrardatos()
print(". ݁₊ ⊹ . ݁˖ . ݁")

print("--- Registro de Pago Numero 2 ---")
pago2 = Pago()
pago2.leerdatos()
pago2.mostrardatos()
print(". ݁₊ ⊹ . ݁˖ . ݁")

print("Software realizado por Zavala Medina Alexia Paola NL = 54+3 = 57")
```
